The BLM_PRODUCT is the key field in identifying the type of activity that the associated Case is designed to establish and moniter. This field is used to group cases by Case Type (commonly referred to as Case Groups), which is the primary method of workflow organization at the BLM. Case Processors (also known as Adjudicators) are responsible for knowing a select set of rules and legislation associated with their Case Type and associated Products, and only a select few Adjudicators are familiar with more than one Case Type. In some scenarios, the Case Type is associated so many Products that Adjudicators are only familiar and trained to process thir specific set of products within the Case Group.
