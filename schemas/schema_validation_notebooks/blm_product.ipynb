From this analysis it is clear that the ID field is the foregn key to the other table BLM_Product codes, and the NAME field is the associated human readable field used to display the Product in the GUI and reports. The other field used in the display is commonly referred to as the "BLM Product Code" and by examining values these are found to be in the CASE_TYPE_CODE field. These two fields (NAME AND CASE_TYPE_CODE) will be used to support the request to remove any erroneous and/or mismatched BLM Product Codes and Names, because currently any reports run on these fields are producing incomplete data as a result of the data mismatches.

In order to export the NAME and CASE_TYPE_CODE fields to perform the review for errors, the two fields must be joined together.


```python
DECLARE snapshot_date STRING DEFAULT '20250901';

EXECUTE IMMEDIATE FORMAT("""
    SELECT 
        IFNULL(MAX(NAME), 'found no match in name') AS NAME,
        CASE_TYPE_CODE
    FROM `blm_seta_dqimp.blm_product_%s`
    GROUP BY CASE_TYPE_CODE
    ORDER BY CASE_TYPE_CODE
""", snapshot_date);
```

This code produces a table with 923 CASE_TYPE_CODES that match NAMES, and for the 40 codes that don't match a NAME the text "found no match in name" was inserted in the table.
