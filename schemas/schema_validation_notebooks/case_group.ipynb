The concept of the Case Group is one that has had ambiguity in the past, and is further exacerbated by the recent migration of LR2000 and the associated merging of the (now deprecated) ALIS system. The main issue with Case Groups stems from the wide variety of BLM Product Names and their associated interpretations. 

As a result, there is no dedicated lookup table for Case Groups. Case Groups have been 'solved' by different users for different needs, and there are a resulting 3 known places to see these attempts. The first is the Record Type ID and the second are is the Case Group field - both stored in the Blm_product__c , along with Case Category, Commodities, Type Code, Legacy System Pcode and Name which might also be useful in this analysis: 
<small> Case_Category__c, Case_Commodities__c, Case_Group__c, Case_Record_Types__c, Case_Type_Code__c, Legacy_System_Code__c, Name,</small>

The other well known effort in classifying BLM Prodyct types into groups to create Case Groups, are the NLSDB feature services. Those services use the product types to generate definition queries that display Case Groups on the MLRS Research Map and in the NLSDB Feature Services for ease of access by Adjudicators and the Public seeking to look at cases and lands directly related to their interest and application process.
