The concept of the Case Type Group is one that has been used by the BLM for many decades. This very common terminology refers to a series of Case Type Codes that have been established to distinguish between Case Processor workflows that are designated by select consumer activities on the land. The Case Type Codes are made human readable through an association to the Case Type Name, however there is known variation in the interpretation of the meaning of the Case Type Name across the different states and their respective operations (ie different workflows). Case Type Groups were created to simplify the different workflow and consumer activities. Case Type Groups are most commonly referred to as Mining Claims, Solid Minerals, Fluid Minerals, Land Use Authorizations, Land Tenure Cases, and Land Transfer Cases (Land Transfer Cases are in Alaska Only). Currently there are over 900 Case Type Codes.

With the creation of MLRS came the introduction of the terminology "BLM Products". BLM Product Codes are synonymous with Case Type Codes, however in the development of the system there has been some ambiguity introduced in the one-to-one relationship between Case Types Codes and BLM Product Codes that still needs to be resolved in order for efficient system operation in the future (and by default, more efficient BLM operations). The conflation of some Case Type Codes and BLM Product Codes is a result of developer error but it has been further exacerbated by the recent migration of LR2000 and the associated merging of the (now deprecated) ALIS system - as many Case Type Codes had different meanings (specific interpretation of process implementation) across different regions. 

Currently, there is no dedicated lookup table for Case Type Groups (aka BLM Product Type Groups). There is active discussion around eliminating any fields in the MLRS systep named "Case Type" so that "BLM Product" can be used reliably as the single source of truth. Until this improvement is implemented, users must take caution when reporting on any information using Product Type Codes, Product Type Names and Product Type Groups.

The Blm_product__c table has all of these ambiguous fields stored in one place. These fields should all be considered inacurate and unreliable when found to be referenced, except for BLM Product Code (aka "ID") and Procuct Name (aka "NAME") and Commodity (aka CAsE_COMMODITIES).

While the MLRS design team works to solve the issue of unreliability in the fields in the Product Table, the NLSDB team has completed an effort involving worksessions with Case Type subject matter experts to classify BLM Prodyct Types into Product Type Groups in order to more intuitively display the NLSDB data and deliver it out in feature services. Those services use the Product Types to generate definition queries that display Case Type Groups (aka Product Type Groups) on the MLRS Research Map and in the NLSDB Feature Services for ease of access by Adjudicators and the Public seeking to look at cases and lands directly related to their interest and application process. As a result of this effort's known reliability, any grouping of data for display and discovery purposes will use the Case Type Groups designated in this analysis and documentation.

The first step in the process is to reverse engineer the Definition Queries used to publish the NLSDB Feature Services. This effort produced a simplified set of sub-types associated with the six main types, all with associated queries designed to select a range of BLM Product Codes and associated BLM Product Names.

The results of these queries were then compared with the output of the full BLM Product table distinct values query. The individual Case Type Group queries were then compared to the output of the full BLM Product table to discover which Product Codes and Names were not in the definition queries. These select few outliers were assigned to a Case Type Group in a worksession with BLM staff. These final queries are stored in this GitHub Repository under schemas>case_type_group_queries as individual queries.
